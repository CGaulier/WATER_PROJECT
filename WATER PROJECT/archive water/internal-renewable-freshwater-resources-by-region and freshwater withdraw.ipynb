```python
df1 = pd.read_csv('data/annual-freshwater-withdrawals.csv') #retrait d'eau douce
df1
```

```python
df1.isna().sum()
```

```python
plt.subplots(figsize=(25,7))
sns.lineplot(x='Year', y='Annual freshwater withdrawals (m³)', hue="Entity", data=df1.nsmallest(20, 'Annual freshwater withdrawals (m³)'))
```

```python
plt.subplots(figsize=(18,7))
sns.lineplot(x='Year',y='Annual freshwater withdrawals (m³)',hue='Entity',data=df1[(df1.Entity == 'United States') | (df1.Entity == 'China') | (df1.Entity == 'Palestine') | (df1.Entity =='Nigeria')|(df1.Entity =='India') | (df1.Entity == 'France')] )
```
